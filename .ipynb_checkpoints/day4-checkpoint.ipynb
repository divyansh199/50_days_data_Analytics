```python
rng = np.random.default_rng(seed=24)
```

```python
arr1 = rng.integers(10,20,(2,3))
arr2 = rng.integers(10,20,(1,3))
```

```python
arr1,arr2
```

```python
arr3 = np.add(arr1,arr2)
```

```python
arr3
```

```python
arr4 = np.dot(arr1,arr2.transpose())
```
