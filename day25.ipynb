```python
import matplotlib.pyplot as plt
```

```python
df_least_5_pro = df4.sort_values(by = "Filter", ascending =True)[0:6]
```

```python
df_least_5_pro
```

```python
# create data
products = df_least_5_pro["Product Name"]
sales = df_least_5_pro["Revenue"]
costs = df_least_5_pro["Costs"]
profit = df_least_5_pro["Filter"]

#figure
fig, ax =plt.subplots(figsize=(12,8))

#ploting the data

ax.bar(products,sales, label = "Sales")
ax.bar(products, costs, label = "Costs", bottom=sales)
ax.bar(products, profit, label = "Filter", bottom= sales+costs)

ax.set_ylabel("Revenue",fontsize =15)
ax.set_xlabel("Products",fontsize =15)
plt.xticks(rotation =0)


ax.legend()
plt.show()
```
