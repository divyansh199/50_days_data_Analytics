```python
df = pd.read_csv("employees_record_data.csv")
```

```python
df
```

```python
import sqlite3
```

```python
engine = sqlite3.connect("Employee_Records.db")
cur = engine.cursor()
```

```python
try:
    df.to_sql("MyEmployeesTable", con=engine, index = False)
except ValueError:
    print("Table already exists")
    
```

```python
df_data = cur.execute("""SELECT * FROM MyEmployeesTable""")
df_data.fetchall()
```

```python
data_table = pd.read_sql_query("""SELECT * FROM MyEmployeesTable""" , con=engine)
data_table
```

```python
salary_over_60k = pd.read_sql_query("""
                                     SELECT Name, Salary
                                     FROM MyEmployeesTable
                                     WHERE Salary > 60000
                                    """, con=engine)
salary_over_60k
```

```python
avg_salary = pd.read_sql_query("""
                                SELECT Department,AVG(Salary) as Average_salary
                                FROM MyEmployeesTable
                                GROUP BY Department
                                """, con = engine)
avg_salary
```

```python
high_salary = pd.read_sql_query("""
                                SELECT Department, Name, MAX(Salary) AS Highest_salary
                                FROM MyEmployeesTable
                                """, con=engine)
high_salary
```

```python
empl_bf_dt = pd.read_sql_query("""
                                SELECT Name FROM MyEmployeesTable
                                WHERE HireDate < "2020-01-01"
                                """, con=engine)
```
