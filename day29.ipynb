```python
df5 = df4.sort_values(by = "Population" , ascending = False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (15,8))

sns.barplot(x = "Country" , y = "Population" , data = df5)
plt.xlabel("Countries")
plt.ylabel("Size")
plt.title("country vs size")
plt.xticks(rotation = 90)
plt.show()
```
