```python
import pandas as pd
```

```python
df = pd.read_csv("website_data_analysis.csv")
```

```python
df["website"].duplicated()
```

```python
avg = df.groupby("website")["visits"].mean()
```

```python
df["uni_visit"] = df["visits"]/df["unique_visitors"]
```

```python
df
```

```python
df2 = df.groupby("website").sum()
```

```python
df2
```

```python
df3 = df2.copy()
```

```python
df3 = df3.sort_values(by="visits",ascending = False).head(5)
df3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x = df3.index , y = "visits" , data = df3)
plt.xlabel("Website")
plt.ylabel("Sites")
# plt.xticks(rotation = 90)
plt.show()
```
