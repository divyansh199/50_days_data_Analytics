```python
nd = np.random.default_rng(seed=42)
arr1=nd.standard_normal((1000))
arr1
```

```python
mn=np.mean(arr1)
mn
```

```python
sd=np.std(arr1)
sd
```

```python
snd=(arr1-mn)/sd
snd
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.hist(snd,bins=20)
plt.title('std normal dist',fontsize=20)
plt.xlabel('x')
plt.ylabel('y')
plt.show
```
